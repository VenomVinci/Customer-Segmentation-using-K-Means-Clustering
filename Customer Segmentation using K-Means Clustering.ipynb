```python
# importing the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# collecting the data

data = pd.read_csv("Mall_Customers.csv")
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# removing columns that are not needed.

X = data.iloc[:,[3, 4]].values

# first values represent annual income, and the second spending score
```

```python
X
```

```python
# choosing number of clusters

# we will be using WCSS( Within-Cluster Sum of Squares )
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init = "k-means++", random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
# elbow graph

sns.set()
plt.plot(range(1, 11), wcss)
plt.title("ELBOW GRAPH")
plt.xlabel("number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
# so from the graph the optimal number of cluster is 5
```

# Training the K-MEANS CLUSTERING model.

```python
kmeans = KMeans(n_clusters = 5, init = "k-means++", random_state=0)
```

```python
Y = kmeans.fit_predict(X)
Y

# this are the predicted clusters
```

```python
# visualizing the clusters

plt.figure(figsize=(8,8))
plt.scatter(X[Y==0,0], X[Y==0,1], s=50, c='red', label='Cluster 1')
plt.scatter(X[Y==1,0], X[Y==1,1], s=50, c='blue', label='Cluster 2')
plt.scatter(X[Y==2,0], X[Y==2,1], s=50, c='purple', label='Cluster 3')
plt.scatter(X[Y==3,0], X[Y==3,1], s=50, c='violet', label='Cluster 4')
plt.scatter(X[Y==4,0], X[Y==4,1], s=50, c='cyan', label='Cluster 5')

# plot the centroids
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='black', label='Centroids')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
